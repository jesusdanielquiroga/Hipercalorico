14- Escriba un programa en python que grafique el comportamiento de la siguiente ecuación:

$x_{x+1}=Rx_{t}(1-x_{t})$

Para la gráfica, considere que esta ecuación tiene un comportamiento caotico y se debe ilustrar su grafica de bifurcación. Puede usar el siguiente código:

```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import pandas as pd

def logistic_iteration(N, xinicial, r):
  logistic=0
  for i in range(N):
    logistic = r*xinicial*(1-xinicial)
    xinicial = logistic
  return logistic
valor_f = []
eje_r = []
valor_r = list(np.linspace(0, 4, 10000))

Xo= 0.5
for i in range (len(valor_r)):
  eje_r.append(valor_r[i])
  valor_f.append(logistic_iteration(len(valor_r), Xo, valor_r[i]))
print(valor_r)
print(eje_r)
print(valor_f)

plt.scatter(eje_r, valor_f)
plt.show()
```
